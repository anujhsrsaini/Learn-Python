```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv(r"data\movies.csv")
movies
```

```python
# Let's extract year from movies
movies['year'] = movies['title'].str.extract(r'\((\d{4})\)', expand=False)
movies['year'] = movies['year'].fillna(1700).astype(int)
```

```python
ratings = pd.read_csv(r"data\ratings.csv")
ratings.head(5)
```

```python
ratings['date_time'] = pd.to_datetime(ratings['timestamp'], unit='s')
ratings
```

```python
ratings.groupby(['movieId']).agg(Mean=('rating', 'mean'), Count=('rating', 'count'))
```

```python
genres = movies.assign(genres=movies['genres'].str.split('|')).explode('genres')
genres
```

```python
genres.groupby(['genres']).agg(Count=('movieId', 'count'))
```
